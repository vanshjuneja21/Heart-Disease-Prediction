Splitting the features and target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

Splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

X_train = pd.get_dummies(X_train,drop_first=True)
X_test = pd.get_dummies(X_test,drop_first=True)

X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

# Store column names for later use in prediction
feature_columns_after_dummies = X_train.columns
```

```python
X_test
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**MODEL TRAINING**

Logistic Regression

```python
lr_model = LogisticRegression(max_iter=1500)
lr_model.fit(X_train,Y_train)

# Predictions
lr_train_prediction = lr_model.predict(X_train)
lr_test_prediction = lr_model.predict(X_test)

#Accuracy
lr_train_data_accuracy = accuracy_score(lr_train_prediction, Y_train)
lr_test_data_accuracy = accuracy_score(lr_test_prediction, Y_test)

print('Accuracy on Training data : ', lr_train_data_accuracy)
print('Accuracy on Test data : ', lr_test_data_accuracy)
```

```python
print("Confusion Matrix:\n", confusion_matrix(Y_test, lr_test_prediction))
print("\nClassification Report:\n", classification_report(Y_test, lr_test_prediction))
```

Decision Tree

```python
dt_model = DecisionTreeClassifier(max_depth=4,random_state=42)
dt_model.fit(X_train, Y_train)

# Predictions
dt_train_prediction = dt_model.predict(X_train)
dt_test_prediction = dt_model.predict(X_test)

# Accuracy
dt_train_data_accuracy = accuracy_score(dt_train_prediction, Y_train)
dt_test_data_accuracy = accuracy_score(dt_test_prediction, Y_test)

print("Accuracy on Training data:", dt_train_data_accuracy)
print("Accuracy on Test data:", dt_test_data_accuracy)
```

```python
print("Confusion Matrix:\n", confusion_matrix(Y_test, dt_test_prediction))
print("\nClassification Report:\n", classification_report(Y_test, dt_test_prediction))
```

Model Evaluation

```python
print("Model Prediction Summary:")
print("Logistic Regression Test Accuracy:",lr_test_data_accuracy)
print("Decision Tree Test Accuracy:",dt_test_data_accuracy)
```
